```python
import math

num_columns = len(data.columns)
rows = cols = math.ceil(num_columns ** 0.5)  

fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(cols * 4, rows * 4))  
axes = axes.flatten()  

for i, column in enumerate(data.columns[:-1]):
    axes[i].hist(data[column], bins=10, color='skyblue', edgecolor='black')
    axes[i].set_title(f"Histogram of {column}")
    axes[i].set_xlabel(column)
    axes[i].set_ylabel("Frequency")


for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python

corr_matrix = (data[data.columns[:-1]]).corr()

# Heatmap 시각화
plt.figure(figsize=(20, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# 제목 추가
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
numerical = ['Application order', 'Age at enrollment','Curricular units 1st sem (credited)',
       'Curricular units 1st sem (enrolled)',
       'Curricular units 1st sem (evaluations)',
       'Curricular units 1st sem (approved)',
       'Curricular units 1st sem (grade)',
       'Curricular units 1st sem (without evaluations)',
       'Curricular units 2nd sem (credited)',
       'Curricular units 2nd sem (enrolled)',
       'Curricular units 2nd sem (evaluations)',
       'Curricular units 2nd sem (approved)',
       'Curricular units 2nd sem (grade)',
       'Curricular units 2nd sem (without evaluations)', 'Unemployment rate', 'Inflation rate', 'GDP']

categorical = []
for col in data.columns[:-1]:
    if col not in numerical:
        categorical.append(col)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np





target_values = sorted(data['Target'].unique())  # 타겟 값 (0, 1, 2)


fig, axes = plt.subplots(nrows=len(numerical) + len(categorical), ncols=len(target_values), figsize=(len(target_values) * 4, (len(numerical) + len(categorical)) * 4))

for row_idx, feature in enumerate(categorical):
    for col_idx, target in enumerate(target_values):
        ax = axes[row_idx, col_idx]  
        sns.histplot(
            data=data[data['Target'] == target], 
            x=feature, 
            ax=ax, 
            discrete=True,  
            color='lightblue', 
            kde=False
        )
        ax.set_title(f"{feature} (Target={target})")
        ax.set_xlabel("")  
        ax.set_ylabel("Count" if col_idx == 0 else "")  

# 🔹 연속형 변수에 대해 히스토그램 그리기
for row_idx, feature in enumerate(numerical, start=len(numerical)):
    for col_idx, target in enumerate(target_values):
        ax = axes[row_idx, col_idx]  
        sns.histplot(
            data=data[data['Target'] == target], 
            x=feature, 
            ax=ax, 
            kde=True, 
            color='lightgreen'
        )
        ax.set_title(f"{feature} (Target={target})")
        ax.set_xlabel("")  
        ax.set_ylabel("Count" if col_idx == 0 else "") 

plt.tight_layout()
plt.show()
```

```python
n_features = len(categorical)
fig, axes = plt.subplots(nrows=n_features, ncols=1, figsize=(6, 6 * n_features))

for i, feature in enumerate(categorical):
    ax = axes[i]
    # 각 카테고리별로 target의 비율을 계산
    target_counts = data.groupby(feature)['Target'].value_counts(normalize=True).unstack()
    
    # stacked bar chart로 표시
    target_counts.plot(kind='bar', stacked=True, ax=ax, colormap='Set2')
    ax.set_title(f"Target Distribution by {feature}")
    ax.set_xlabel(feature)
    ax.set_ylabel("Proportion")
    ax.legend(title='Target', bbox_to_anchor=(1, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
